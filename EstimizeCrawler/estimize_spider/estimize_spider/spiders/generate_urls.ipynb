```python
import pandas as pd
```

```python
tickers_df = pd.read_excel('data/tickers.xlsx')
tickers = tickers_df['ticker'].apply(lambda x: str(x).lower()).tolist()
```

```python
def gen_link(ticker, fq, year):
    main = 'https://www.estimize.com/'
    return main + ticker + '/' + 'fq' + str(fq) + '-' + str(year) + '?metric_name=eps&chart=historical'
```

```python
urls = []
count = 0
for year in range(2012, 2020):
    for fq in range(4):
        fq += 1
        for ticker in tickers:
            urls.append(gen_link(ticker, fq, year))
            print(count, ticker, year, fq)
            count += 1
        if year == 2019 and fq == 1:
            break
```

```python
df = pd.DataFrame(columns=['url'])
```

```python
df['url'] = urls
```

```python
df.loc[0:41658].to_excel('urls_1.xlsx', encoding='UTF-8', index=False)
```
